```python
import pandas as pd
```

```python
df= pd.read_csv('cleaned_unemployment.csv')
```

```python
df.head()
```

```python
# Remove leading and trailing whitespace from the column names
df.columns = df.columns.str.strip()
```

```python
# Check for missing values
print(df.isnull())
```

```python
print(df.isnull().sum())
```

```python
#Check for duplicates
print(df.duplicated().sum())
```

```python
#Check current data types
print(df.dtypes)
```

```python
#filter rows based on a condition
filtered_df=df[df['Unemployment Rate']>70]
print(filtered_df)
```

```python
#sort a DataFrame
#Sort by one column
df.sort_values('Unemployment Rate',ascending=True)
```

```python
## Average unemployment by Region
avg_unemployment_rate =df.groupby('Region')['Unemployment Rate'].mean()
```

```python
print(avg_unemployment_rate)
```

```python
avg_unemployment =df['Unemployment Rate'].mean()
```

```python
print(avg_unemployment)
```

```python
# Add a new column for average 
df['Average Unemployment Rate'] = avg_unemployment
```

```python
highest_unemployment_rate = df['Unemployment Rate'].max()
print(highest_unemployment_rate)
```

```python
df['Highest Unemployment Rate'] = highest_unemployment_rate 
```

```python
total_labour_participatiion = df['Labour Participation Rate'].sum()
```

```python
print(total_labour_participatiion )
```

```python
df['Total Labour Rate'] = total_labour_participatiion 
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# visualize data in Python using matplotlib
df['Unemployment Rate'].hist()
plt.title('Unemployment')
plt.show()
```

```python
# visualize data in Python using seaborn

sns.barplot(x='Unemployment Rate',y='Area',data=df)
plt.show()
```

```python
plt.figure(figsize=(10,9))
sns.barplot(x='Unemployment Rate',y='Region',data=df)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Unemployment Rate', data=df)
plt.title('Unemployment Rate Over Time in India')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Regional Unemployment Rate Comparison Purpose: To compare the unemployment rate across different regions at a specific point in time.
plt.figure(figsize=(35,6))
sns.barplot(x='Region',y='Unemployment Rate',data=df)
plt.title('Unemployment Rate by Region in India')
plt.xlabel('region')
plt.ylabel('unemployment Rate(%)')
plt.tight_layout()
plt.show()
```

```python
# Group data by region and calculate the average unemployment rate
regional_avg = df.groupby('Region')['Unemployment Rate'].mean().reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Unemployment Rate', data=regional_avg)
plt.title('Average Unemployment Rate by Region in India')
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Create pivot table
pivot_table = df.pivot_table(values='Unemployment Rate',index='Region',aggfunc='sum')
print(pivot_table)
```
